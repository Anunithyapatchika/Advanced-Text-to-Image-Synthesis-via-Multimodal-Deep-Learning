```python
# Install required packages
!pip install torch torchvision transformers diffusers gradio

# Import necessary libraries
import torch
from transformers import CLIPModel, CLIPTokenizer
from diffusers import DiffusionPipeline
import gradio as gr

# Initialize the device
device = "cuda" if torch.cuda.is_available() else "cpu"

# Load the Latent Consistency Model
pipe = DiffusionPipeline.from_pretrained("SimianLuo/LCM_Dreamshaper_v7")
pipe.to(device=device, dtype=torch.float32)  # Opt for float32 for better precision

# Load CLIP for text encoding
clip_model = CLIPModel.from_pretrained("openai/clip-vit-base-patch32")
clip_tokenizer = CLIPTokenizer.from_pretrained("openai/clip-vit-base-patch32")
clip_model.to(device)

def encode_text(prompt):
    # Encode text using CLIP and return features
    inputs = clip_tokenizer(prompt, return_tensors="pt", truncation=True, padding=True).to(device)
    with torch.no_grad():
        text_features = clip_model.get_text_features(**inputs)
    return text_features

def generate_image_from_text(prompt, num_inference_steps=4, guidance_scale=8.0):
    # First encode the text to features
    text_features = encode_text(prompt)
    # Generate images using the LCM model
    images = pipe(prompt=prompt, num_inference_steps=num_inference_steps, guidance_scale=guidance_scale, output_type="pil").images
    return images

# Define a Gradio interface to interact with the model
def gradio_interface(prompt):
    images = generate_image_from_text(prompt)
    return images[0]  # Return the first image from the batch

# Setup Gradio interface
iface = gr.Interface(fn=gradio_interface, inputs="text", outputs="image", title="Text-to-Image Synthesis using LCM")
iface.launch()
```
